```python
import pandas as pd
import numpy as np
```

```python
from transformers import pipeline
```

```python
sentiment_pipeline = pipeline('sentiment-analysis', model='nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
data = {
    'text': [
        'Eu adoro usar a biblioteca transformers da hugging face!',
        'Estou achando tudo horroroso',
        'Não sei o que dizer, mas demorou pra chegar a informação',
        'Estou decepcionado com o serviço',
        'Achei esse lugar muito sujo'
    ]
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
def analyze_sentiment(text):
  result = sentiment_pipeline(text)[0]
  label = result['label']
  score = result['score']

  if label == '1 star':
    sentiment = 'negative'
  elif label == '2 stars' or label == '2 stars':
    sentiment = 'neutral'
  elif label == '4 stars' or label == '5 stars':
    sentiment = 'positive'
  else:
    sentiment = 'neutral' #caso qualquer outro rótulo
  return sentiment, score
```

```python
df[['sentiment', 'score']] = df['text'].apply(lambda x: pd.Series(analyze_sentiment(x)))
```
